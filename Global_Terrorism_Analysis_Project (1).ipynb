# Data Vizualization, Storytelling & Experimenting with charts : Understand the relationships between variables

### Analyze the trends over time and identify any long-term patterns or fluctuations in terrorist activities.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the GTD dataset
gtd_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Global Terrorism Data.csv', encoding='latin-1')
# Convert 'iyear' column to datetime format
gtd_data['iyear'] = pd.to_datetime(gtd_data['iyear'], format='%Y')

# Group the data by year and count the number of attacks for each year
attacks_by_year = gtd_data.groupby(gtd_data['iyear'].dt.year).size()

# Create a line chart to visualize the frequency of terrorist attacks by year
plt.figure(figsize=(10, 6))
plt.plot(attacks_by_year.index, attacks_by_year.values, marker='o', linestyle='-')
plt.title('Terrorist Attacks by Year')
plt.xlabel('Year')
plt.ylabel('Number of Attacks')
plt.grid(True)
plt.show()
```

###  Compare the distribution of attacks across different categories and identify any hotspots or regions with higher levels of activity.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the Global Terrorism Database
gtd_data = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Global Terrorism Data.csv', encoding='latin-1')

# Identify categorical variable (e.g., country or region)
categorical_variable = 'country_txt'

# Group data by the categorical variable and count the number of attacks for each category
attacks_by_category = gtd_data.groupby(categorical_variable).size().sort_values(ascending=False)

# Plot the frequency of attacks for each category using a bar chart
plt.figure(figsize=(12, 6))
attacks_by_category.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries with Highest Number of Terrorist Attacks')
plt.xlabel('Country')
plt.ylabel('Number of Attacks')
plt.xticks(rotation=45, ha='right')
plt.show()

```
